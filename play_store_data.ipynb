# Importing Libraries

```python
import pandas as pd       
import matplotlib.pyplot as plt 
import seaborn as sns
import numpy as np
```

# Importing dataset

```python
data=pd.read_csv("E:\\archive (2)\\googleplaystore.csv")
```

```python
data
```

```python
data.info()
```

```python
data.describe()
```

# Checking The Null Values

```python
data.isna().sum()
```

# Dropping The Un necessary Colomns

```python
data.drop(['Current Ver','Genres','Last Updated','Android Ver'],axis=1,inplace=True)
data
```

```python
data.info()
```

# Filling The Missing Values 

```python
data['Rating']=data['Rating'].ffill()
data.dropna(how ='any', inplace = True)
data
```

```python
plt.figure(figsize=(20,10))
g = sns.kdeplot(data.Rating,color='g',shade = True)
g.set_xlabel("Rating")
g.set_ylabel("Frequency")
plt.title('Distribution of Rating',size = 20)
plt.show()
```

# Managing The Reviews

```python
data['Reviews']=data['Reviews'].str.replace('.0M','000')
data['Reviews']=data['Reviews'].astype('int64')
```

```python
data
```

# Managing the Price

```python
data['Price']=data['Price'].str.replace('$','')
data['Price']=data['Price'].str.replace('Everyone','0')
data['Price']=data['Price'].astype('float64')
```

```python
data_paid=data.loc[data['Price']!=0]
data_paid
```

```python
data_free=data.loc[data['Price']==0]
data_free
```

```python
data['Types'] = data['Type'].map(lambda i :0  if i =='Free' else 1)
data['Types'].value_counts().plot(kind='pie',autopct='%1.1f%%')
plt.title('Freee v/s Paid Apps',size = 20)
plt.show()
```

# Managing The Installs

```python
data['Installs']=data['Installs'].str.replace('+','')
data['Installs']=data['Installs'].str.replace(',','')
data['Installs']=data['Installs'].astype('int64')
```

```python
data
```

# Managing The Size 

```python
data['Size']=data['Size'].str.replace('k','1000')
data['Size']=data['Size'].str.replace('M','1000000')
data['Size'].replace('Varies with device', np.nan, inplace = True )
data['Size']=data['Size'].fillna(data['Size'].median())
data['Size']=data['Size'].astype('float64')
```

```python
data
```

# Distribution of Rating v/s Category

```python
plt.figure(figsize=(25,13))
sns.boxplot(x='Category',y='Rating',data=data)
plt.title('Distribution of Rating v/s Category',size = 20)
plt.xticks(rotation=90)
plt.show()
```

# Apps with Category

```python
plt.figure(figsize=(20,15))
sns.countplot(x='Category',data=data)
plt.xticks(rotation=90)
plt.title('Category Counts',size=20)
plt.show()
```

# Distribution of Category v/s Reviews

```python
plt.figure(figsize=(20,10))
sns.barplot(x='Category',y='Reviews',data=data)
plt.xticks(rotation=90)
plt.title('Distribution of Category v/s Reviews',size=20)
plt.show()
```

# Category v/s Installs

```python
plt.figure(figsize=(20,10))
sns.barplot(x='Category',y='Installs',data=data)
plt.xticks(rotation=90)
plt.title('Distribution of Category v/s Installs',size=20)
plt.show()
```

# Size v/s Installs

```python
plt.figure(figsize=(20,10))
sns.lineplot(x='Size',y='Installs',data=data,color='g')
plt.xticks(rotation=90)
plt.title('Distribution of Size v/s Reviews',size=20)
plt.show()
```

# Category v/s Size 

```python
plt.figure(figsize=(20,10))
sns.lineplot(x='Category',y='Size',data=data,color='g')
plt.title('Distribution of Category v/s Size',size=20)
plt.xticks(rotation=90)
plt.show()
```

# Content rating v/s Installs 

```python
plt.figure(figsize=(20,10))
sns.pointplot(x='Content Rating',y='Installs',data=data,color='g')
plt.xticks(rotation=90)
plt.title('Distribution of Content Rating v/s Installs',size=20)
plt.show()
```
